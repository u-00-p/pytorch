## Introduction to Tensors

```python
# scalar
scalar = torch.tensor(7)
scalar
```

```python
#Dimension (number of brackets)
print(scalar.ndim)
# Get tensor back as python int
print(scalar.item())
```

```python
# Vector
vector = torch.tensor([7,7])
vector
```

```python
#Dimension
print(vector.ndim)
print(vector.shape)
```

```python
# Matrix
matrix = torch.tensor([[7,8],
                       [9,10]])
matrix
```

```python
print(matrix.ndim)
print(matrix.shape)
print()
print(matrix[0])
print(matrix[1])
print()
print(matrix[0][0])
print(matrix[0][1])
print(matrix[1][0])
print(matrix[1][1])
print()
print(matrix[0,0])
print(matrix[0,1])
print(matrix[1,0])
print(matrix[1,1])
```

```python
# tensor
tensor = torch.tensor([[[1,2,3,4,5],
                        [4,5,6,7,8],
                        [7,8,9,10,11]]])
tensor
```

```python
print(tensor.ndim)
print(tensor.shape)
```
