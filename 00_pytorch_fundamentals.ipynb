## Introduction to Tensors

```python
# scalar
scalar = torch.tensor(7)
scalar
```

```python
#Dimension (number of brackets)
print(scalar.ndim)
# Get tensor back as python int
print(scalar.item())
```

```python
# Vector
vector = torch.tensor([7,7])
vector
```

```python
#Dimension
print(vector.ndim)
print(vector.shape)
```

```python
# Matrix
MATRIX = torch.tensor([[7,8],
                       [9,10]])
MATRIX
```

```python
print(MATRIX.ndim)
print(MATRIX.shape)
print()
print(MATRIX[0])
print(MATRIX[1])
print()
print(MATRIX[0][0])
print(MATRIX[0][1])
print(MATRIX[1][0])
print(MATRIX[1][1])
print()
print(MATRIX[0,0])
print(MATRIX[0,1])
print(MATRIX[1,0])
print(MATRIX[1,1])
```

```python
# tensor
TENSOR = torch.tensor([[[1,2,3,4,5],
                        [4,5,6,7,8],
                        [7,8,9,10,11]]])
TENSOR
```

```python
print(TENSOR.ndim)
print(TENSOR.shape)
```

```python
TENSOR[0]
```

```python
TENSOR[0,0]
```

```python
TENSOR[0,0,0]
```

### Random tensors

```python
random_tensor = torch.rand(3,4,2)
random_tensor
```

```python
print(random_tensor.ndim)
print(random_tensor.shape)
```

```python
random_tensor = torch.rand(1,2,3,3,3)
random_tensor
```

```python
print(random_tensor.ndim)
print(random_tensor.shape)
```

```python
# similar tensor image shape
random_image_size_tensor = torch.rand(3,224,244) #colour channels,height, width
print(random_image_size_tensor.ndim)
print(random_image_size_tensor.shape)
```
